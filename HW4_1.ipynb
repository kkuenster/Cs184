```python
# HW 4 Part 1
# Part A: Scipy
# step 1
# Load the dataset from ‘Wholesale customers data.csv’ into a pandas DataFrame
# and print it along with its shape.

import pandas as pd

df = pd.read_csv('Wholesale customers data.csv')

print(df)

df.shape
```

```python
# step 2 
# Normalize the whole dataset using the normalize() function. [Note: Output of
# normalize() is a numpy array]

from sklearn import preprocessing
import numpy as np

normalized_arr = preprocessing.normalize(df)
print(normalized_arr)
```

```python
# step 3

# import scipy.cluster.hierarchy as shc. Use the following template to print the
# dendrogram for your dataset
# a. plt.figure(figsize=(N, M))  – N and M are the size of the plot.
# b. plt.title("Enter title of your dendrogram")
# c. dend = shc.dendrogram(shc.linkage(your data variable, method=XXX))
# Find out what the ‘method’ parameter is and select the ones covered in class (single, complete, average)

import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

plt.figure(figsize=(50,20))
plt.title('Dendrogram for df')
dend = shc.dendrogram(shc.linkage(normalized_arr, method = 'single'))
```

```python
plt.figure(figsize=(50,20))
plt.title('Dendrogram for df')
dend = shc.dendrogram(shc.linkage(normalized_arr, method = 'complete'))
```

```python
plt.figure(figsize=(50,20))
plt.title('Dendrogram for df')
dend = shc.dendrogram(shc.linkage(normalized_arr, method = 'average'))
```
