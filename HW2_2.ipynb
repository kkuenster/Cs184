```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn import tree

df=pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv", names=["Pregnancies", "Glucose", "BloodPressure", "SkinThickness","Insulin", "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"])
print(df)
print(df.shape)
x=df.drop("Outcome", axis=1)
y=df['Outcome']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)
for mDepth in range(2,11):
    print("\nMax Depth: ", mDepth)
    pidTree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=mDepth, random_state=1) #ID3 uses entropy
    pidTree.fit(x_train, y_train)  # your model
    trainPrediction = pidTree.predict(x_train)
    print("Train Accuracy: ", metrics.accuracy_score(y_train, trainPrediction))
    testPrediction = pidTree.predict(x_test)
    print("Test Accuracy: ", metrics.accuracy_score(y_test, testPrediction))
```
