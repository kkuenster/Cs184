```python
# step 4: perform label encoding on target variable
from sklearn import preprocessing

l_e = preprocessing.LabelEncoder()

df['diagnosis'] = l_e.fit_transform(df['diagnosis'])
print(df.head())
```

```python
# step 5: divide data into x: features and y: target 
import numpy as np

x = df.drop(['diagnosis','id'], axis=1)
y = df['diagnosis']

# checking for normality 
df['diagnosis'].value_counts()

# normalizing data 
# we were not sure if this was necessary so we commented it out
#x = (x - np.min(x)) / (np.max(x) - np.min(x))
```

```python
# step 6: Split data into test and train with 70-30 split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 44)
```

```python
# step 7: naive bayes

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
```

```python
# step 8: model performance 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

predictions = nb.predict(x_test)
y_scores = nb.predict_proba(x_test)

# Get evaluation metrics
cm = confusion_matrix(y_test, predictions)
print ('Confusion Matrix:\n',cm, '\n')
print('Accuracy:', accuracy_score(y_test, predictions))
print("Overall Precision:",precision_score(y_test, predictions))
print("Overall Recall:",recall_score(y_test, predictions))
print("F1 Score:", f1_score(y_test,predictions))
auc = roc_auc_score(y_test,y_scores[:,1])
print('AUC: ' + str(auc))
```
