```python
import pandas as pd

# load dataset
df = pd.read_csv('train.csv')

# returns dataset and shape of set
print(df)
df.shape
```

```python
# checking for null columns 
print(df.isnull().sum())
```

```python
# dropping any column that contains 1 or more NaN
df=df.dropna(axis='columns')
```

```python
# printing new shape shows 19 columns were dropped 
print(df.shape)
```

```python
# remove 'ID' column
df.drop(columns=['Id'])
```

```python
# finding columns with < 10 unique values
unique = df.nunique()
print(unique)
```
