```python
# split data into X and y
# step 6 part a
# numerical features only
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# print(df.dtypes)

# features
#X = df.drop(['SalePrice','Condition1','Condition2','SaleType','SaleCondition','PavedDrive', 
                    #'Functional','HeatingQC','Foundation','CentralAir','Heating','ExterCond',
                    #'ExterQual','RoofStyle','RoofMatl','HouseStyle','BldgType','LandSlope', 
                    #'LotConfig','LandContour','LotShape','Street','MSZoning'], axis=1)
# target
#Y = df['SalePrice']

# create test and train groups 

#x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 44)

```

```python
# step 6 part b
# one hot encoding 
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

# shows all rows and columns
pd.set_option('max_rows', None)
pd.set_option('max_columns', None)

# defining vars to be encoded
categorical_cols = ['Condition1','Condition2','SaleType','SaleCondition','PavedDrive', 
                    'Functional','HeatingQC','Foundation','CentralAir','Heating','ExterCond',
                    'ExterQual','RoofStyle','RoofMatl','HouseStyle','BldgType','LandSlope', 
                    'LotConfig','LandContour','LotShape','Street','MSZoning']

# encoding 
transformer = make_column_transformer(
    (OneHotEncoder(), ['Condition1','Condition2','SaleType','SaleCondition','PavedDrive', 
                    'Functional','HeatingQC','Foundation','CentralAir','Heating','ExterCond',
                    'ExterQual','RoofStyle','RoofMatl','HouseStyle','BldgType','LandSlope', 
                    'LotConfig','LandContour','LotShape','Street','MSZoning']),
    remainder='passthrough')

# fitting and combining with original df
transformed = transformer.fit_transform(df)
transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names())
print(transformed_df.head())

print(transformed_df)
```

```python
# training one hot encoded model


X = transformed_df.drop(['SalePrice'], axis=1)
Y = transformed_df['SalePrice']

# create test and train groups 

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 44)

reg = LinearRegression()
```

```python
from sklearn.metrics import mean_absolute_error
# using one hot encoded categorical and numerical features 
reg.fit(x_train,y_train)
predictions = reg.predict(x_test)
mae_two = mean_absolute_error(y_test, predictions)
print('the error is {}'.format(mae_two))
```
